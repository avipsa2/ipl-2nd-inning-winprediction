```python
match_df=match_df[['match_id','total_runs','city','winner']]
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df['balls_left']=126-(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')
delivery_df['player_dismissed'].unique()
```

```python
delivery_df['current_runrate']=delivery_df['current_score']*6/(120-delivery_df['balls_left'])
```

```python
delivery_df['req_runrate']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
wickets=delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
```

```python
delivery_df['wickets_left']=10-wickets
```

```python
delivery_df.head()
```

```python
def resultfun(x):
    return 1 if x['batting_team']==x['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(resultfun,axis=1)
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','balls_left','wickets_left','total_runs_x','current_runrate','req_runrate','result']]
```

```python
final_df
```

```python
final_df.isna().sum()
```

```python
final_df=final_df.fillna(0)
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
categorical_cols = ['batting_team', 'bowling_team', 'city']
for col in categorical_cols:
    final_df[col] = final_df[col].astype(str)
```

```python
final=final_df.copy()
test=final['result']
train=final.drop('result',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(train,test,test_size=0.3,random_state=1)
```

```python
from sklearn import metrics
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier
```

```python
cf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
final
```

```python

pipe = Pipeline(steps=[
    ('step1',cf),
    ('step2',LogisticRegression(solver='liblinear'))
])

pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
pipe.predict_proba(X_test)[5]
```

```python
import xgboost as xgb
X_encoded = pd.get_dummies(train, columns=['batting_team', 'bowling_team', 'city'])
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, test, test_size=0.3, random_state=42)
# Initialize the XGBoost regressor
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Train the regressor
xgb_reg.fit(X_train, y_train)
# Make predictions
y_pred = xgb_reg.predict(X_test)
# Evaluate the regressor
mse = metrics.r2_score(y_test, y_pred)
print(f"XGBoost Mean Squared Error: {mse:.2f}")
```
